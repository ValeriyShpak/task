## 1. Сколько можно сделать шагов максимально/минимально при утилизации ягод в разрезе разных данных

## В разрезе справочных данных:

### Минимальное количество шагов будет сделано в случае, если мы будем утилизировать только волчьи ягоды:

```python
data_1_min = in_forest[in_forest['steps'] < 0]

math.floor((data_1_min['steps'] * data_1_min['count']).sum())
```

### Максимальное количество шагов будет сделано, если мы будем утилизировать все ягоды кроме волчьих:

```python
data_1_max = in_forest[in_forest['steps'] > 0]

math.floor((data_1_max['steps'] * data_1_max['count']).sum())
```

## В разрезе данных о найденных ягодах:

### Дополним таблицу 'find_in_forest' данными о количестве шагов для каждой ягоды, при этом:
- волчьи ягоды будут давать -18 шагов; 
- остальные ягоды будут давать максимально возможное для своей массы количество шагов.
- согласно справочным данным, количество шагов имеет понятную зависимость от массы ягоды только для ежевики. Найденная ежевика имеет массу 5, 6 и 7. Проведя интерполяцию на основании 2 точек из таблицы 'in_forest', получим соответствующее количество шагов для ягод ежевики различной массы: 5g-19steps, 6g-15steps, 7g-11steps.

```python
def determine_steps(x):
    
    if x['berry'] == 'волчья ягода':
    
        steps = -18
        
    elif x['berry'] == 'ежевика':
        
        if x['weight'] == 5:
        
            steps = 19
            
        elif x['weight'] == 6:
            
            steps = 15
            
        else:
            
            steps = 11
            
    elif x['berry'] == 'брусника':
    
        steps = 20
        
    elif x['berry'] == 'малина':
        
        steps = 19
        
    elif x['berry'] == 'крыжовник':
        
        steps = 6
        
    elif x['berry'] == 'арбуз':
        
        steps = 900
        
    else:
        
        steps = 18
        
    return steps

find_in_forest['steps'] = find_in_forest.apply(lambda x: determine_steps(x), axis=1)
```

### Минимальное количество шагов:

```python
data_2_min = find_in_forest[find_in_forest['steps'] < 0]

math.floor(data_2_min['steps'].sum())
```

### Максимальное количество шагов:

```python
data_2_max = find_in_forest[find_in_forest['steps'] > 0]

math.floor(data_2_max['steps'].sum())
```

## 5. Найти соотношение количества ягод из выгрузки, при котором будет сделано 0 шагов

### Подготовим и отобразим датафрейм для анализа:

```python
grouped_find_in_forest_5 = find_in_forest.groupby(['berry', 'weight', 'steps'])['date'].count().to_frame().reset_index()

grouped_find_in_forest_5.rename(columns = {'date':'count'}, inplace=True)

grouped_find_in_forest_5_2 = grouped_find_in_forest_5.groupby(['berry', 'steps']).agg({'count' : 'sum',
                                                                                       'weight' : 'first'}).reset_index()

grouped_find_in_forest_5_2
```

### Определим соотношения ягод из выгрузки, при которых будет сделано 0 шагов (для количества ягод не более 4):

```python
# Найдём диапазоны значений для каждого варианта:

grouped_find_in_forest_5_2['range']=grouped_find_in_forest_5_2.apply(lambda x: 
                                                                     [i for i in range(0, x['steps']*x['count'], x['steps'])], 
                                                                     axis=1)

count = 1

# Найдём соотношения:

print('\033[1m' + 'При данных комбинациях будет сделано 0 шагов:' + '\033[0m')

print()

for x0 in grouped_find_in_forest_5_2['range'][0][:5]:
    
    for x1 in grouped_find_in_forest_5_2['range'][1][:5]:
            
            for x2 in grouped_find_in_forest_5_2['range'][2][:5]:
                
                for x3 in grouped_find_in_forest_5_2['range'][3][:5]:
                    
                    for x4 in grouped_find_in_forest_5_2['range'][4][:5]:
                        
                        for x5 in grouped_find_in_forest_5_2['range'][5][:5]:
                            
                            for x6 in grouped_find_in_forest_5_2['range'][6][:5]:
                                    
                                    if x0 + x1 + x2 + x3 + x4 + x5 + x6 == 0:
                                        
                                        x_list = [x0, x1, x2, x3, x4, x5, x6]
                                        
                                        print('\033[1m' + 'Комбинация номер ' + str(count) + ':' + '\033[0m')
                                        
                                        for i in range(len(grouped_find_in_forest_5_2)):
                                            
                                            print(grouped_find_in_forest_5_2['berry'][i] + ' (' 
                                                  + str(grouped_find_in_forest_5_2['weight'][i]) + ' г.): ' 
                                                  + str(int(x_list[i]/grouped_find_in_forest_5_2['steps'][i])) + ' шт.')
                                        
                                        print()
                                        
                                        count += 1
```
